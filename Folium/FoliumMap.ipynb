```python
import folium

# Definindo as coordenadas de Porto Alegre
latitude = -30.0346
longitude = -51.2177

# Criando um mapa centrado em Porto Alegre
mapa = folium.Map(location=[latitude, longitude], zoom_start=12)

# Adicionando um marcador para Porto Alegre 
folium.Marker([latitude, longitude], popup='Porto Alegre, RS, Brasil').add_to(mapa)

# Exibindo o mapa
mapa
```

```python
import folium

# Definindo as coordenas das capitais brasileiras:
# Mapa do Brasil (Região Sul)
latitude = -33.4507
longitude = -53.2350

# Criando um mapa centrado em Porto Alegre
mapa = folium.Map(location=[latitude, longitude], zoom_start=3)

# Porto Alegre
latitude = -30.0346
longitude = -51.2177

# Adicionando um marcador para Porto Alegre
folium.Marker([latitude, longitude], popup='Porto Alegre, RS, Brasil').add_to(mapa)

# Exibindo o mapa
mapa

# São Paulo
latitude = -23.5505
longitude = -46.6333

# Adicionando um marcador para São Paulo
folium.Marker([latitude, longitude], popup='São Paulo, SP, Brasil').add_to(mapa)

# Exibindo o mapa
mapa

# Rio de Janeiro
latitude = -22.9068
longitude = -42.1729

# Adicionando um marcador para Rio de Janeiro
folium.Marker([latitude, longitude], popup='Rio de Janeiro, RJ, Brasil').add_to(mapa)

# Exibindo o mapa
mapa

# Curitiba
latitude = -25.4290
longitude = -49.2671

# Adicionando um marcador para Curitiba
folium.Marker([latitude, longitude], popup='Curitiba, PR, Brasil').add_to(mapa)

# Exibindo o mapa
mapa

# Belo Horizonte
latitude = -19.9173
longitude = -43.9345

# Adicionando um marcador para Belo Horizonte
folium.Marker([latitude, longitude], popup='Belo Horizonte, MG, Brasil').add_to(mapa)

# Exibindo o mapa
mapa
```
