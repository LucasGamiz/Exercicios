```python
import pandas as pd

# Carregar o arquivo CSV em um DataFrame
file_path = 'archive/fipe_cars.csv'
fipe_df = pd.read_csv(file_path)

# Mostrar quantos 'year_of_reference' diferentes existem
unique_years = fipe_df['year_of_reference'].nunique()
print(f"Quantos 'year_of_reference' diferentes existem: {unique_years}")

# Contar quantas entradas existem para cada ano de referência
entries_per_year = fipe_df['year_of_reference'].value_counts()
print("Número de entradas para cada ano de referência:")
print(entries_per_year)
```

```python
import pandas as pd

# Carregar o arquivo CSV em um DataFrame
file_path = 'archive/fipe_cars.csv'
fipe_df = pd.read_csv(file_path)

# Filtrar os dados para o ano de referência 2023
fipe_df_2023 = fipe_df[fipe_df['year_of_reference'] == 2023]

# Verificar se há dados para 2023 antes de calcular
if not fipe_df_2023.empty:
    # Encontrar o preço MÍNIMO e MÁXIMO em 2023
    preco_min_2023 = fipe_df_2023['avg_price_brl'].min()
    preco_max_2023 = fipe_df_2023['avg_price_brl'].max()

    # Exibir os resultados
    print(f"Preço MÍNIMO de um carro em 2023: R$ {preco_min_2023:.2f}")
    print(f"Preço MÁXIMO de um carro em 2023: R$ {preco_max_2023:.2f}")
else:
    print("Não há dados disponíveis para o ano de referência 2023.")
```

```python
import pandas as pd

# Carregar o arquivo CSV em um DataFrame
file_path = 'archive/fipe_cars.csv'
fipe_df = pd.read_csv(file_path)

# Filtrar os dados para carros com preço entre R$ 5.000 e R$ 7.000
carros_na_faixa = fipe_df[(fipe_df['avg_price_brl'] >= 5000) & (fipe_df['avg_price_brl'] <= 7000)]

# Verificar se há carros na faixa de preço
if not carros_na_faixa.empty:
    # Selecionar as colunas 'year_of_reference', 'model', e 'avg_price_brl'
    modelos_precos_ano = carros_na_faixa[['year_of_reference', 'model', 'avg_price_brl']]
    
    # Ordenar os carros por 'avg_price_brl' em ordem crescente
    modelos_precos_ano_ordenados = modelos_precos_ano.sort_values(by='avg_price_brl')
    
    # Contar o número de carros listados
    num_carros = modelos_precos_ano_ordenados.shape[0]
    
    # Exibir os resultados
    print("Carros que você pode comprar entre R$ 5.000 e R$ 7.000 (ordenados por preço):")
    print(modelos_precos_ano_ordenados.to_string(index=False))
    print(f"\nNúmero de carros listados: {num_carros}")
else:
    print("Não há carros disponíveis entre R$ 5.000 e R$ 7.000.")
```
